```python
def longestCommonPrefix(strs):
    """
    :type strs: List[str]
    :rtype: str
    """
    if len(strs) == 0:
        print ("")
    current = strs[0]
    for i in range(1,len(strs)):
        temp = ""
        if len(current) == 0:
            break
        for j in range(len(strs[i])):
            if j<len(current) and current[j] == strs[i][j]:
                temp+=current[j]
            else:
                break
            current = temp
        return current
```

```python
def longestCommonPrefix(my_str):
    if my_str == []:
        return ''
    if len(my_str) == 1:
        return my_str[0]
    my_str.sort()
    shortest = my_str[0]
    prefix = ''
    for i in range(len(shortest)):
        if my_str[len(my_str) - 1][i] == shortest[i]:
            prefix += my_str[len(my_str) - 1][i]
        else:
            break
    return prefix

my_list_1 = ['car', 'carbon', 'vehicle']
my_list_2 = ['car', 'carbon', 'carbonmonoxide']

# this will give an emptry string
print(longestCommonPrefix(my_list_1))
# this will return the string 'car'
print(longestCommonPrefix(my_list_2))
```

```python
def StudentsLunchRefined(students, sandwiches):
    
    if len(students) == len(sandwiches):
        length = len(students)
        while True:
            if length > 0:
                length = len(students)
                i = students[0]
                j = sandwiches[0]
            
                if i==j:
                    students.pop(0)
                    sandwiches.pop(0)
                    length = length-1
                else:
                    nlist1 = [True for x in students if x==1]
                    nlist2 = [False for x in students if x==0]
                    if length == 0 or len(nlist1) == length or len(nlist2) == length:
                        return length
                        break
                    else:
                        students.insert(length,i)
                        students.pop(0)
            else:
                return length
                break
```
